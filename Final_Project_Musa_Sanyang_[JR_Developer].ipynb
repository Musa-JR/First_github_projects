```python
# showing the departure time and arrival time for flight with their counts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Load the dataframe
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])


plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(df['departure_time'].value_counts().index, df['departure_time'].value_counts().values, color=['r','b'])
plt.title('Departure Time')
plt.xlabel('D Time')
plt.ylabel('D Freq')

plt.subplot(1,2,2)
plt.bar(df['arrival_time'].value_counts().index, df['arrival_time'].value_counts().values, color=['g','y'])
plt.title('Arrival Time')
plt.xlabel('D Time')
plt.ylabel('D Freq')


plt.show()
```

```python
#Q.3.showing the Source City & Destination City with their counts
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.barh(df['source_city'].value_counts().index, df['source_city'].value_counts().values , color=['y','g'])
plt.title('Source City')
plt.xlabel('No.Flights')
plt.ylabel('City')

plt.subplot(1,2,2)
plt.barh(df['destination_city'].value_counts().index, df['destination_city'].value_counts().values , color=['m','r'])
plt.title('Destination City')
plt.xlabel('No.Flights')
plt.ylabel('City')

plt.show()
```
